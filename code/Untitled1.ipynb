```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename = '20180713.csv'
df = pd.read_csv(f'..//Data//metroKOR//{filename}')
print(df.columns)
```

```python
for word in ['Room', 'Outside']:
    df[f'Train {word} Temp ＃1'] = df[f'Train {word} Temp ＃1'].str[:-3]
    df[f'Train {word} Temp ＃1'] = df[f'Train {word} Temp ＃1'].astype('int64')
    df.rename({f'Train {word} Temp ＃1': f'Train {word} Temp ＃1 (℃)'}, axis=1, inplace=True)

threewords = ['P/B', 'DISTANCE', 'Line Voltage', 'Distance to Target']
for word in threewords:
    df[f'{word}'] = df[f'{word}'].str[:-3]
    df[f'{word}'] = df[f'{word}'].astype('int64')
    
df.rename({'P/B': 'P/B (%)'}, axis=1, inplace=True)
df.rename({'DISTANCE': 'DISTANCE (m)', 'Distance to Target': 'Distance to Target (m)'}, axis=1, inplace=True)
df.rename({'Line Voltage': 'Line Voltage (V)'}, axis=1, inplace=True)
    
for i in [1, 2, 3, 4, 7, 0]:
    df[f'BC ＃{i}'] = df[f'BC ＃{i}'].str[:-5]
    df[f'BC ＃{i}'] = df[f'BC ＃{i}'].astype('int64')
    df.rename({f'BC ＃{i}': f'BC ＃{i} (kPa)'}, axis=1, inplace=True)
    
df['MR Pressure'] = df['MR Pressure'].str[:-5]
df['MR Pressure'] = df['MR Pressure'].str.replace('．', '.')
df['MR Pressure'] = df['MR Pressure'].astype('float64')
df.rename({'MR Pressure': 'MR Pressure (bar)'}, axis=1, inplace=True)

for word in ['Target', 'Permitted', 'Actual', 'Train']:
    df[f'{word} Speed'] = df[f'{word} Speed'].str[:-6]
    df[f'{word} Speed'] = df[f'{word} Speed'].astype('int64')
    df.rename({f'{word} Speed': f'{word} Speed (km/h)'}, axis=1, inplace=True)
    
df.drop(['Unnamed: 27', 'Unnamed: 28'], axis=1, inplace=True)
```

```python
for ind, word in enumerate(['Target', 'Permitted', 'Actual']):
    df.insert(12+ind, f'{word} Speed (cm/s)', 0)
    df[f'{word} Speed (cm/s)'] = df[f'{word} Speed (km/h)'] * 27.778
    df[f'{word} Speed (cm/s)'] = df[f'{word} Speed (cm/s)'].astype('float64')
    
df.insert(17, 'Train Speed (cm/s)', 0)
df['Train Speed (cm/s)'] = df['Train Speed (km/h)'] * 27.778
df['Train Speed (cm/s)'] = df['Train Speed (cm/s)'].astype('float64')
    
```
